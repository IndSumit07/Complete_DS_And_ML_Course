# Functions
- A function is a block of code that performs a specific task.

```python
#Syntax
def function_name(parameters):
    """DocString"""
    #Function body
    return expression
```

```python
def even_or_odd(num):
    """This function finds if a number is even or odd"""
    if(num%2==0):
        print("Even")
    else:
        print("Odd")
```

```python
#calling a function
even_or_odd(10)
```

```python
#functions with multiple parameters
def add(a, b):
    return a+b

print(add(2,4))
```

```python
#Deafult parameters
def greet(name="Someone"):
    print(f"Hello {name}")

greet()
```

```python
#variable length arguments
#positional and keywords arguments

def print_numbers(*args):
    for i in args:
        print(i)

print_numbers(1,2,3,4,5, "Sumit", "Mathura")
```

```python
#Positional arguments
def print_numbers(*args):
    for i in args:
        print(i)
print_numbers(1,2,3,4,5, "Sumit", "Mathura")
```

```python
#Keywords arguments
def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="Sumit", age=19, city="Basti")
```

```python
#Keywords arguments
def print_details(*args, **kwargs):

    for val in args:
        print(f"Positional arguments: {val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")

print_details(1,2,3,4,5, name="Sumit", age=19, city="Basti")
```

```python
#Return statements
def multiply_divide(a,b):
    return a*b, a/b

print(multiply_divide(3,2))
```
