# Dictionary
- Dictionaries are unordered collections of items.
- They store data in key-value pairs.
- Keys must be unique and immutable
- Dictionaries are mutable

```python
#Creating dictionaries
empty_dict = {}
print(type(empty_dict))

empty_dict = dict()
```

```python
student = {"name": "Sumit", "age": 19, "salary": 10000000.0}
student["city"]="Basti"
print(student)
```

```python
#Accessing dict elements
print(student["name"])
print(student["age"])
print(student["salary"])
```

```python
#Accessing using get() method
print(student.get("city"))
```

```python
print(student.get('last_name', "Not Available"))
```

```python
#Modifying dict elements
student["name"] = "Sumit Kumar"
print(student)
```

```python
#Deleting a key
del student["city"]
student
```

```python
#Dictionary methods
keys = student.keys()      #get all the keys
print(keys)

values = student.values()  #get all the values
print(values)

items=student.items()      #get all the key value pairs -> list of tupples
print(items)
```

# Shallow Copy

```python
student_copy = student
print(student)
print(student_copy)
```

```python
student_copy["name"]="Uttam"
print(student_copy)
print(student)
```

```python
student_copy1=student.copy()
student_copy1["name"]="Rahul"
print(student_copy1)
print(student)
```

# Iterating over dictionaries

```python

#you can use loops to iterate over dict, keys, values or items

for keys in student.keys():
    print(keys)
```

```python
for values in student.values():
    print(values)
```

```python
for key,value in student.items():
    print(f"{key}:{value}")
```

# Nested Dictionaries

```python
students = {
    "student1": {"name":"Sumit", "age":19},
    "student2": {"name":"Uttam", "age":20},
    "student3": {"name":"Shivang", "age":18},
}
```

```python
#Access nested dictionaries elements
print(students["student1"]["name"])
print(students["student2"]["age"])
```

```python
#Iterating over nested dict
for student_id, student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
#dictionaries comprehension
squares = {x:x*x for x in range(5)}
squares
```

```python
#conditional dictionaries comprehension
evens={x:x*x for x in range(10) if x%2==0}
evens
```

```python
#Practical examples
numbers = [1,1,1,1,2,2,2,2,4,4,6,7,8,1,5,2,2]
frequency = {}

for i in numbers:
    if i in frequency:
        frequency[i]+=1
    else:
        frequency[i]=1

frequency
```

```python
#merge dictionaries
dict1 = {"a":1, "b":2}
dict2 = {"b":3, "c":4}
merged_dict={**dict1, **dict2}
merged_dict
```
