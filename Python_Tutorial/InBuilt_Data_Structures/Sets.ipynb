# Sets
- Sets are built-in data type in python used to store collections of unique items.
- They are unordered means that the elements do not follow a specific order and they do not allow duplicate elemets.
- Sets are useful for membership tests, eliminating duplicates entries and performing mathematical set operations like union, intersection, difference and symmetric difference.

```python
#Creating sets
my_set = {1,2,3,4,5}
print(my_set) 
```
