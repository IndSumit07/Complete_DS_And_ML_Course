# List Comprehension
 - List Comprehension is a powerful and concise way to create lists in python.
 - They are syntactically compact and can replace more verbose looping constructs.

```python
lst = [x**2 for x in range(15)]
lst
```

#### Basic Syntax               [expression for item in iterable]

#### With Conditional Logic     [expression for item in iterable if condition]

#### Nested List Comprehension  [expression for item1 in iterable for item2 in interable]

```python
#Basic List Comprehension
lst = [i**2 for i in range(10)]
print(lst)
```

```python
#List Comprehension With Condition
square_of_even_nums = [i**2 for i in range(10) if i%2==0]
print(square_of_even_nums)
```

```python
#Nested List Comprehension
lst1=[1,2,3,4]
lst2=['a', 'b', 'c', 'd']

pair=[[item1, item2] for item1 in lst1 for item2 in lst2]
pair
```

```python
#List Comprehension with function calls
phones = ["Iphone", "Samsung", "Realme", "Nothing", "Redmi"]
count_chars = [len(phone) for phone in phones]
count_chars
```
