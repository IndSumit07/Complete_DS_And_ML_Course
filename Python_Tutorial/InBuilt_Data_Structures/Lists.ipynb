# Lists are ordered, mutable collections of items.
- They can contain items of different data types.

```python
#Empty list
lst = []
print(type(lst))
```

```python
names = ["Sumit", "Shivang", "Uttam", 1, 2, 3]
print(names)
```

```python
mixed_list = [1, "Hello", 3.14, True]
print(mixed_list)
```

```python
#Accessing List Elemets
fruits = ["Apple", "Mango", "Banana", "Grapes", "Strawberry"]

for i in range (0, len(fruits)):
    print(fruits[i])

print(fruits[-1])
```

```python
print(fruits[1:])
print(fruits[1:3])
print(fruits[:])
print(fruits[::])
```
