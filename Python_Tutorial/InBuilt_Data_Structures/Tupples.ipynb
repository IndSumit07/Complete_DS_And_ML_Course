# Tupples
- Tuples are ordered collection of items that are immutable.
- They are similiar to lists, but their immutability makes them different.

```python
#Creating a Tuple
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
#Creating empty lists and tuple
lst = list()
tup = tuple()
```

```python
numbers = (1,2,3,4,5)
print(numbers)
```

```python
tuple_from_list = ([1,2,3,4,5])
list_from_tuple = [(1,2,3,4,5)]
tuple_from_list
list_from_tuple
```

```python
mixed_tuple = (1, "Sumit", 3.14, True)
mixed_tuple
```

```python
#Accessing tuple elements
numbers[0]
numbers[1]
```

```python
numbers[0:4]
numbers[::-1]
```

# Tuple Operations

```python
concatinated_tuple = numbers+mixed_tuple
concatinated_tuple
```

```python
mixed_tuple*3
numbers*5
```

# Tuple Methods

```python
print(numbers.count(1))
print(numbers.index(3))
```

# Paciking and Unpacking Tuples

```python
packed_tuples = 1, "hello", 3.14
print(packed_tuples)
```

```python
a,b,c = packed_tuples
print(a)
print(b)
print(c)
```

```python
numbers = (1,2,3,4,5,6)
first, *middle, last = numbers
print(first)
print(middle)
print(last)
```
